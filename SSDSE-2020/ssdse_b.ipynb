```python
#----検証12.2006年から2017年について，降水量（年間）が最も多い都道府県はそれぞれどこだろう.----
data_b2 = data_b.groupby(["都道府県","年度"]).sum().reset_index()
```

```python
year = range(2006,2017,1)
for x in year:
    filename = 'precipitation-%s.csv' % x
    data_b2_1 = data_b2[data_b2["年度"] == x]
    data_b2_2 = data_b2_1[["都道府県","年度","降水量（年間）"]]
    data_b2_3 = data_b2_2.sort_values("降水量（年間）")
    data_b2_3.to_csv(filename)
#----終了----
```
