```python
data_a = pd.read_csv("SSDSE-2020A.csv", encoding="shift-jis", header=2)
```

```python
data_a.head()
```

```python
#すべてのカラム情報
data_a.info()
```

```python
#----検証1.人口が10,000人未満の市区町村はいくつあるのか.----
data_a1=data_a[data_a["総人口"]<10000]
data_a1[["市区町村","総人口"]]
```

```python
len(data_a1)
#----終了----
```

```python
#----検証2.人口が10,000人未満の市区町村は，中国5県（鳥取，島根，山口，広島，岡山）において合計いくつあるのか.----
data_a2 = data_a[(data_a["総人口"]<10000) & (data_a["都道府県"].isin(["鳥取県","島根県","山口県","広島県","岡山県"]))]
data_a2[["都道府県","市区町村","総人口"]]
```

```python
len(data_a2)
#----終了----
```

```python
#----検証3.中学校数が一つしかない市区町村はいくつあるのか.----
data_a3 = data_a[(data_a["中学校数"]==1)]
data_a3[["都道府県","市区町村","中学校数"]]
```

```python
len(data_a3)
#----終了----
```

```python
#----検証4.小学校も中学校もそれぞれ一つしかない市区町村はいくつあるのか.----
data_a4 = data_a[(data_a["小学校数"] == 1) & (data_a["中学校数"] == 1)]
data_a4[["都道府県","市区町村","小学校数","中学校数"]]
```

```python
len(data_a4)
#----終了----
```

```python
#----検証5.高校がない市区町村はいくつあるのか.----
data_a5 = data_a[data_a["高等学校数"]==0]
data_a5[["都道府県","市区町村","高等学校数"]]
```

```python
len(data_a5)
#----終了----
```

```python
#----検証6.全世帯に占める高齢単身世帯数（65歳以上の者１人）の割合が多い市区町村の上位3つとその割合はなんだろうか.----
data_a["全世帯に占める高齢単身世帯数の割合"] = data_a["高齢単身世帯数（65歳以上の者１人）"] / data_a["世帯数"] 
data_a6 = data_a.sort_values("全世帯に占める高齢単身世帯数の割合", ascending = False)
data_a6[["都道府県","市区町村","世帯数","高齢単身世帯数（65歳以上の者１人）","全世帯に占める高齢単身世帯数の割合"]]
```

```python
data_a6[["都道府県","市区町村","世帯数","高齢単身世帯数（65歳以上の者１人）","全世帯に占める高齢単身世帯数の割合"]][0:3]
#----終了----
```

```python
#----検証7.人口が10番目，20番目，30番目に多い都道府県とその人口は何人だろう----
data_a7 = data_a.groupby("都道府県").sum()
data_a7_1 = data_a7.sort_values("総人口", ascending =False)
```

```python
data_a7_1[["総人口"]][9:30]
#----終了----
```

```python
#----検証8.転出者数（日本人移動者）が多い都道府県の上位3つとその数はいくつだろう.----
data_a8 = data_a.groupby("都道府県").sum().reset_index()
data_a8_1 = data_a.sort_values("転出者数（日本人移動者）", ascending =False)
```

```python
data_a8_1[["都道府県","転出者数（日本人移動者）"]][0:3] 
#----終了----
```
